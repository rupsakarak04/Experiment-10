```python
# Create DataFrame
data = { "Name": ["A","B","C"], "Marks": [90,80,78]}
df = pd.DataFrame(data)
df
```

```python
display(df.head())  #prints first 5 rows
df.tail()  #prints last 5 rows
```

```python
# Find the structure of the DataFrame
print(df.shape)

# Find Dimension
print(df.ndim)

# Find no of elements in the table
print(df.size)

# Display column names
print(df.columns)
```

```python
# Check Data types of each column
df.dtypes
```

```python
# Access Individual column
display(df["Name"])
df["Marks"]
```

```python
# Access specific row
df.loc[1]
```

```python
# Displaying a specific row, column
df.iloc[1,1]
```

```python
# Add a new column
df["Grade"] = ["First Class", "Distinction", "Second Class"]
df
```

```python
# Update Data in DataFrame
# Change marks of Student A
df.loc[0, "Marks"] = 98
df
```

```python
#Delete a column
df.drop("Grade", axis = 1, inplace = False)
display(df)

df.drop("Grade", axis = 1, inplace = True)
display(df)
```

```python
# Performing Basic Statistical operations

mean = df["Marks"].mean()
display(mean)

min = df["Marks"].min()
display(min)

max = df["Marks"].max()
display(max)
```

```python
# Apply Condition(Filtering)
df[df["Marks"]> 80]
```
